## Show a breakdown of distance from home by job role and attrition.

```python
df1 = df[['DistanceFromHome','Education','JobRole','Attrition','MonthlyIncome']]
```

```python
dff= df[['Attrition','DistanceFromHome','JobRole']]
dff
```

```python
Attrition = pd.get_dummies(dff.Attrition, prefix = 'Attrition', prefix_sep = '_')
Attrition
```

```python
df2 = pd.concat([dff,Attrition ], axis = 1)
df2 = df2.drop("Attrition",1)
df2
```

```python
df3 = df2.groupby(['DistanceFromHome','JobRole']).sum()
df3
```

## Compare average monthly income by education and attrition.

```python
df6 = df[['Education','Attrition','MonthlyIncome']]
df6
```

```python
df7 = df6.groupby(['Education', 'Attrition']).mean()
df7
```
